## Embedding Model

```python
from llama_index.embeddings.huggingface import HuggingFaceEmbedding
embed_model=HuggingFaceEmbedding(model_name="sentence-transformers/all-mpnet-base-v2")
```

## Prompt

```python
from llama_index.core import PromptTemplate
```

```python
from llama_index.llms.huggingface import HuggingFaceLLM
```

```python
system_prompt=""""<|SYSTEM|># You are a Q&A assistant. Your goal is to answer questions as
accurately as possible based on the instructions and context provided

"""
```

```python
# This will wrap the default prompts that are internal to llama-index
query_wrapper_prompt = PromptTemplate("<|USER|>{query_str}<|ASSISTANT|>")
```

```python
from huggingface_hub import notebook_login
notebook_login()
```

## Create LLM Wrapper-Mistral

```python
# https://huggingface.co/mistralai/Mistral-7B-Instruct-v0.1
import torch
llm=HuggingFaceLLM(
    context_window=4096,
    max_new_tokens=256,
    generate_kwargs={"temperature": 0.7, "do_sample": False},
    system_prompt=system_prompt,
    query_wrapper_prompt=query_wrapper_prompt,
    tokenizer_name="mistralai/Mistral-7B-Instruct-v0.1",
    model_name="mistralai/Mistral-7B-Instruct-v0.1",
    device_map="auto",
    stopping_ids=[50278, 50279, 50277, 1, 0],
    tokenizer_kwargs={"max_length": 4096},
    # uncomment this if using CUDA to reduce memory usage
    model_kwargs={"torch_dtype": torch.float16}

)
```

## Chunkins

```python
from llama_index.core import ServiceContext

Service_Context=ServiceContext.from_defaults(
    chunk_size=1024,
    llm=llm,
    embed_model=embed_model
)
```

## llama_index- VectorStoreIndex

```python
from llama_index.core import VectorStoreIndex
```

```python
index=VectorStoreIndex.from_documents(documents,service_context=Service_Context)
```

```python
query_engine=index.as_query_engine(k=1)
```

```python
result=query_engine.query("which matrics we can use for evaluating the machine translation task?")
```
